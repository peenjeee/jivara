### 5.2 Assessing Data

Data ini berupa teks naratif tidak terstruktur. Kita perlu mengekstrak tingkat keparahan menggunakan aturan Regex (Rule-Based NLP).

```python
# Sampel interaksi untuk memahami pola teks
sample_texts = []
for drug in drug_interactions_raw[:10]:
    for text in drug.get('food_interactions', []):
        sample_texts.append({'drug': drug['name'], 'text': text[:100]})

pd.DataFrame(sample_texts)
```

### 5.3 Cleaning & Feature Extraction — NLP Severity Scoring

Mengekstrak tingkat keparahan dari teks naratif menggunakan pola Regex:
- **AVOID_CRITICAL (4)**: Mengandung kata "avoid", "fatal", "contraindicate"
- **MONITOR (3)**: Mengandung kata "monitor", "increase", "decrease"
- **LIMIT (3)**: Mengandung kata "limit", "reduce"
- **INTAKE_TIMING (2)**: Mengandung kata "timing", "meal", "food"
- **INFO_MODERATE (1)**: Tidak cocok dengan pola di atas

```python
def classify_interaction_severity(text):
    """Klasifikasi keparahan interaksi obat-makanan dari teks naratif."""
    text_lower = text.lower()

    if re.search(r'\bavoid\b|\bfatal\b|\bcontraindicate\b', text_lower):
        return 'AVOID_CRITICAL', 4
    elif re.search(r'\bmonitor\b|\bincrease\b|\bdecrease\b', text_lower):
        return 'MONITOR', 3
    elif re.search(r'\blimit\b|\breduce\b', text_lower):
        return 'LIMIT', 3
    elif re.search(r'\btim\w*\b|\bmeal\b|\bfood\b', text_lower):
        return 'INTAKE_TIMING', 2
    else:
        return 'INFO_MODERATE', 1

# Proses seluruh dataset
global_knowledge_base = {}
severity_distribution = {'AVOID_CRITICAL': 0, 'MONITOR': 0, 'LIMIT': 0, 'INTAKE_TIMING': 0, 'INFO_MODERATE': 0}

for drug_entry in drug_interactions_raw:
    drug_name = drug_entry.get('name')
    raw_interactions = drug_entry.get('food_interactions', [])
    structured_interactions = []

    for text in raw_interactions:
        interaction_type, severity_score = classify_interaction_severity(text)
        severity_distribution[interaction_type] += 1
        structured_interactions.append({
            'raw_text': text,
            'type': interaction_type,
            'severity_score': severity_score
        })

    if structured_interactions:
        global_knowledge_base[drug_name] = structured_interactions

print(f"Obat dengan interaksi terstruktur: {len(global_knowledge_base)}")
print(f"\nDistribusi severity:")
for severity_type, count in severity_distribution.items():
    print(f"  {severity_type:<18}: {count}")
```

## 8. Feature Engineering

Membuat fitur turunan: persentase kontribusi makronutrien dan kategori makro.

```python
total_macro_calories = (unified_nutrition['proteins'] * 4
                      + unified_nutrition['fat'] * 9
                      + unified_nutrition['carbohydrate'] * 4)
total_macro_calories = total_macro_calories.replace(0, np.nan)

unified_nutrition['pct_protein'] = round((unified_nutrition['proteins'] * 4 / total_macro_calories) * 100, 1)
unified_nutrition['pct_fat'] = round((unified_nutrition['fat'] * 9 / total_macro_calories) * 100, 1)
unified_nutrition['pct_carb'] = round((unified_nutrition['carbohydrate'] * 4 / total_macro_calories) * 100, 1)
unified_nutrition['calorie_density'] = round(unified_nutrition['calories'] / unified_nutrition['weight_grams'], 2)

def classify_macro_category(row):
    if pd.isna(row['pct_protein']):
        return 'Unknown'
    if row['pct_protein'] >= 40:
        return 'High Protein'
    elif row['pct_fat'] >= 50:
        return 'High Fat'
    elif row['pct_carb'] >= 65:
        return 'High Carb'
    return 'Balanced'

unified_nutrition['macro_category'] = unified_nutrition.apply(classify_macro_category, axis=1)

print("Distribusi kategori makro:")
print(unified_nutrition['macro_category'].value_counts())
```
